```
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
%matplotlib inline

df = sb.load_dataset('tips')
sb.regplot(x = 'total_bill', y = 'tip', data = df)
plt.xlabel=('Total Bill')
plt.ylabel=('Bill Tips')
plt.show()
```
